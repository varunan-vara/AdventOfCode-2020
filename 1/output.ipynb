```python
#imports
import re


instructionspart1 = """
Before you leave, the Elves in accounting just need you to fix your expense report (your puzzle input); apparently, something isn't quite adding up.

Specifically, they need you to find the two entries that sum to 2020 and then multiply those two numbers together.

For example, suppose your expense report contained the following:

1721
979
366
299
675
1456

In this list, the two entries that sum to 2020 are 1721 and 299. Multiplying them together produces 1721 * 299 = 514579, so the correct answer is 514579.

Of course, your expense report is much larger. Find the two entries that sum to 2020; what do you get if you multiply them together?

To begin, get your puzzle input.
"""
```

```python
#Read lines 
file_handle = open("input.txt","r")
#Put files from text file into list
_numberlist = file_handle.readlines()
#Create new list
numberlist = []
#Convert string into int
for numbers in _numberlist:
    #Regex - organize
    pattern = r"\n"
    numberstr = re.sub(pattern, "", numbers)
    number = int(numberstr)
    numberlist.append(number)
numberlist
```

```python
#Find length of list
length = len(numberlist)

#Counters to keep track of number of times counted through list
count1 = 0
count2 = 0

#While counters and If statements to test if addition = 2020
while count1 < length - 1:
    while count2 < length - 1:
        newt = numberlist[count1] + numberlist[count2]
        if (newt == 2020):
            print(newt)
            print(numberlist[count1] * numberlist[count2])
        count2 += 1
    count2 = 0
    count1 += 1
```

```python
#THEREFORE the answer is 921504
```
